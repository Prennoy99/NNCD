This notebook demonstrates the workflow of using a Physics-Informed Neural Network (PINN) to model suspension dynamics. The core logic has been extracted into the `src/` directory for modularity and best practices.

````python
import pandas as pd
import matplotlib.pyplot as plt

from src.data_generation import generate_basic_data, generate_robust_episodes, generate_unseen_data
from src.train import train_model
from src.evaluate import evaluate_model

%matplotlib inline
````

## 2. Data Generation
Generate the datasets for training and testing. The dynamics are simulated using the mathematical state-space equations.

````python
# Generate ground truth, robust episodes, and unseen test data
generate_basic_data()
generate_robust_episodes()
generate_unseen_data()
````

## 3. Training the PINN
Train the neural network to mimic the quarter-car physics. The training loop computes both a data loss (against numerical derivatives) and a physics loss (against the state-space matrices).

````python
# Train with Physics-Informed Loss
train_model(use_physics_loss=True, epochs=500, lr=0.005)
````

## 4. Evaluation
Load the trained weights and simulate the quarter-car model over an unseen road profile.

````python
# Evaluate against unseen data
evaluate_model(
    model_path='model_pinn.pth', 
    data_path='unseen_road_test_data_increased_amp.csv', 
    output_img='evaluation_unseen.png'
)
````
